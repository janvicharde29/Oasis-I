```python
data[' Date'] = pd.to_datetime(data[' Date'])
data.dtypes
```

```python
sns.set_style("darkgrid")
sns.pairplot(data,hue = "Area")
```

```python
sns.pairplot(data=data[data['Region'] == 'Andhra Pradesh'], hue='Area')
plt.show() 
```

```python
correlation_matrix = data.corr()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
```

```python
area_counts = data['Area'].value_counts()

plt.figure(figsize=(4, 4))
plt.pie(area_counts, labels=area_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Area')
plt.axis('equal') 
plt.show()
```

```python
data.plot(x=' Date', y=' Estimated Unemployment Rate (%)', figsize=(8,4))
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```
